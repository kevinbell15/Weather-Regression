```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Import WeatherData 
weather_data = pd.read_csv(r"C:\Users\kevinB12\Documents\Project Data\WeatherData.csv") 
# Display data
weather_data
```

```python
# Set Temperature to x
x = weather_data['Temperature(C)']
# Set Humidity to Y
y = weather_data['Humidity']
# Plot the data
plt.scatter(x, y)
# Add labels to the graph
plt.xlabel('Temperature(C)')
plt.ylabel('Humidity')
```

```python
from sklearn.linear_model import LinearRegression
# Change x from 1D array to a 2D array
x = weather_data['Temperature(C)'].values.reshape(-1,1)
y = weather_data['Humidity']
# Define the variable lr_model 
lr_model = LinearRegression()
# Fit the linear regression model to data
lr_model.fit(x, y)
# Find predicted values for all x values 
y_pred = lr_model.predict(x)
plt.scatter(x, y)
plt.xlabel('Temperature(C)')
plt.ylabel('Humidity')
# Plot the regression line
plt.plot(x, y_pred)
```

```python
# Put 32 into the regression model and turn into 2D array
y_pred = lr_model.predict(np.array([32]).reshape(1, 1))
y_pred
```
